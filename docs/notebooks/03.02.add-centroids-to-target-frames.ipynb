```python
import geopandas as gpd
import pandas as pd
```

```python
df = gpd.read_file("../data/pph_fpc_target_s1.shp")
df.to_crs(3587, inplace=True)
```

```python
gdf_points = df.copy()
```

```python
gdf_points['geometry'] = gdf_points['geometry'].centroid
```

```python
gdf_points = gpd.GeoDataFrame(gdf_points, geometry='geometry', crs=df.crs)
```

```python
gdf_points.to_crs(4326, inplace=True)
```
